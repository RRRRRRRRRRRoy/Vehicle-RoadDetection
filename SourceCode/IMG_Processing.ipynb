```python
# This is the number of the images
# From the supp file the images are start from 1
# Setting the counter to 1
counter = 1
```

```python
# HOW TO READ IMAGE BY USING OPENCV IN CHINESE DIRECTORY 
# Source: https://www.programmersought.com/article/3090516575/
def cv_imread(file_path):
    cv_img = cv2.imdecode(np.fromfile(file_path,dtype=np.uint8),-1)
    return cv_img
```

```python
# Read data in batches
# Using the tqdm can check the current process
# Source: https://pypi.org/project/tqdm/
from tqdm import tqdm
data_size = len(data)
for index in tqdm(range(data_size)):
    # If you want to generating the image in macOS, please change this 
    root = 'H:\\Data\\benchmark_velocity_supp\\'
    # Getting the image path
    image_path = root+data[index]['file_name']
    # This print is to check the file dir is correct or not
    print(image_path)

    # read the image by using cv2 ----> previous forget
    # This is used in win10 
    image = cv_imread(image_path)
    # print(type(image))
    # If you are using macOS, Plz use this one
    # image = cv2.imread(image_path)
    # print("image:",image)
    # plt.imshow(image)

    data_bbox = data[index]['bbox']
    # print(len(data_bbox))
    for j in range(len(data_bbox)):
        # !!!!!!!!!!!! careful the data we get is string 
        # report : TypeError: string indices must be integers
        # need change the type
        # This part of code is from the github
        # Source: https://github.com/TuSimple/tusimple-benchmark/blob/master/evaluate/velocity.py
        top = int(data_bbox[j]['top'])
        left = int(data_bbox[j]['left'])
        right = int(data_bbox[j]['right'])
        bottom = int(data_bbox[j]['bottom'])
        rectangled_image = image[top:bottom,left:right]
        # Due to the amount of data is too large
        # omit those picture smaller than 30 * 30 
        length_image = rectangled_image.shape[0]
        width_image = rectangled_image.shape[1]
        if length_image > 30 :
            if width_image > 30:
                vehicle_path = vehicle_output_location + "IMG_" + str(counter) + '.png'
                # check the numpy has no false no zero ndarray.any
                # https://beginnersbook.com/2019/03/python-any-function/
                if rectangled_image.any():
                    # resize to 64 * 64 
                    new_size = (64,64)
                    resized_image = cv2.resize(rectangled_image,new_size)
                    # This is for win10 chinese directory solution of saving vehicles image
                    # Source: https://www.programmersought.com/article/3090516575/
                    cv2.imencode(f"IMG_ + str(counter) + .png",resized_image)[1].tofile(vehicle_path)
                    # U can use this one for macos
                    # This one in win10 can also use but not resized
                    # cv2.imwrite(vehicle_path,rectangled_image)

                    # generating the the non-vehicles roi
                    # This is used the random function to compute the relative distance
                    length_of_image = image.shape[0]
                    width_of_image = image.shape[1]
                    # range of the window ----> except the resize 64 *64 area
                    range_of_length = length_of_image -64
                    range_of_width = width_of_image -64
                    # replace the 64*64 part of the image
                    relative_x = randint(0 , range_of_length)
                    relative_y = randint(0 , range_of_width)
                    # the range of length is from x to x+64
                    # the range of width is from y to y+64
                    nonvehicle_image = image[relative_x:relative_x+64,relative_y:relative_y+64]
                    # similar with the rectangle_image in previous
                    if nonvehicle_image.any():
                        nonvehicle_path = nonvehicle_output_location + "IMG_" + str(counter) + '.png'
                        cv2.imencode(f"IMG_ + str(counter) + .png",nonvehicle_image)[1].tofile(nonvehicle_path)
                        # U can use the following one for macos
                        # This one in win10 can also use but not resized
                        # cv2.imwrite(vehicle_path,rectangled_image)
                counter += 1
```
